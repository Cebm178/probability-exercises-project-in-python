## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# TODO
def estimate_probability(num_simulations=1000):

    dice1 = np.random.randint(1, 7, num_simulations)
    dice2 = np.random.randint(1, 7, num_simulations)

    total_scores = dice1 + dice2

    successful_outcomes = np.sum((total_scores % 2 == 0) | (total_scores > 7))

    probability = successful_outcomes / num_simulations

    return probability

estimated_probability = estimate_probability()
print(f"The estimated probability is: {estimated_probability}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. Take all of the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

# Create the box of balls
for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

print(ball_box)
            
# TODO
```

```python
ball_box = list(ball_box.values())

def draw_balls(ball_box, num_draws=5):
    return [random.choice(ball_box) for _ in range(num_draws)]

def run_experiment(total_experiments=1000):
    count_3_white_2_red = 0
    count_all_same_color = 0

    for _ in range(total_experiments):
        draw = draw_balls(ball_box)
        white_count = draw.count("White")
        red_count = draw.count("Red")
        green_count = draw.count("Green")

        if white_count == 3 and red_count == 2:
            count_3_white_2_red += 1

        if white_count == 5 or red_count == 5 or green_count == 5:
            count_all_same_color += 1

    prob_3_white_2_red = count_3_white_2_red / total_experiments
    prob_all_same_color = count_all_same_color / total_experiments

    return prob_3_white_2_red, prob_all_same_color

prob_3_white_2_red, prob_all_same_color = run_experiment()

print(f"Probability of drawing exactly 3 white and 2 red balls: {prob_3_white_2_red:.4f}")
print(f"Probability of drawing 5 balls of the same color: {prob_all_same_color:.4f}")
```
